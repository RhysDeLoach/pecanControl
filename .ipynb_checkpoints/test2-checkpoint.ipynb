```python
from ultralytics import YOLO 
import cv2
import cvzone
import time

device = 'mps'

model = YOLO("yolo11m-pecan.pt")  # Load your YOLOv8 model

cy1=350
offset=90
idDict={}
pecanCount = 0
badRes = []
frameCount = 0

# Define the desired resolution and FPS
width, height = 1280, 720

cap = cv2.VideoCapture(0)

# Set the resolution and other properties for each camera
cap.set(cv2.CAP_PROP_FRAME_WIDTH, width)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, height)
```

```python
startTime = time.time()

while True:
    ret,frame = cap.read()
    if not ret:
        break

    # Define the black box position (adjust as needed)
    top_left = (0, 0) 
    bottom_right = (1280, 220)  

    # Draw a black rectangle (filled)
    cv2.rectangle(frame, top_left, bottom_right, (0, 0, 0), thickness=-1)
    
    frameCount += 1

    if frameCount % 2 == 0:
    
        # Define the black box position (adjust as needed)
        top_left = (0, 0) 
        bottom_right = (1280, 220)  

        # Draw a black rectangle (filled)
        cv2.rectangle(frame, top_left, bottom_right, (0, 0, 0), thickness=-1)

        results = model.track(frame, stream=True, persist=True,classes = 0, device = device)  

        # Iterate over the results (since results is now a generator)
        for result in results:
            # Check if there are any boxes in the result
            if result.boxes is not None and result.boxes.id is not None:
            # Get the boxes (x, y, w, h), class IDs, track IDs, and confidences
                boxes = result.boxes.xyxy.int().cpu().tolist()  # Bounding boxes
                track_ids = result.boxes.id.int().cpu().tolist()  # Track IDs
                # Iterate over each detected object
                for box, track_id in zip(boxes, track_ids):
                    x1, y1, x2, y2 = box
                    cy = int((y1 + y2) // 2)  # Center Y-coordinate
            
                    # Check if the track ID already exists in the dictionary
                    if track_id in idDict.keys():
                        cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 255, 255), 2)
                        cvzone.putTextRect(frame, f'{idDict[track_id]}', (x1, y2), 1, 1)
                    else:
                        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                        cvzone.putTextRect(frame, f'{track_id}', (x1, y2), 1, 1)

                    # Check if the Y-coordinate of the object is within a certain range
                    if cy < (cy1 + offset) and cy > (cy1 - offset) and track_id not in idDict.keys():
                        pecanCount += 1
                        idDict[track_id] = pecanCount
                        print(pecanCount)

    # Display the total pecan count on the frame
    cv2.putText(frame, f'Totals Pecans: {pecanCount}', (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 
                    0.6, [255, 255, 255], 1)

    # Draw lines on the frame (for visual reference)
    cv2.line(frame, (0, cy1), (1280, cy1), (255, 0, 255), 2)
    cv2.line(frame, (0, cy1 + offset), (1280, cy1 + offset), (0, 0, 255), 2)
    cv2.line(frame, (0, cy1 - offset), (1280, cy1 - offset), (0, 0, 255), 2)

    if frame.shape != (720,1280,3):
        # print(fq"Frame shape: {frame.shape}")  # (height, width, channels)
        badRes.append(frame.shape)
    
    cv2.imshow('Frame', frame)
    if cv2.waitKey(1) & 0xFF == ord("q"):
       break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
print(pecanCount)
print(badRes)
print(frameCount / (time.time()-startTime))
```
