```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from filterpy.kalman import KalmanFilter
import math
```

```python
# Kalman Filter
def kalmanFilter(reading):
    # Initialize Kalman filter
    kf = KalmanFilter(dim_x=2, dim_z=1)

    kf.F = np.array([[1, 1],  # State transition matrix (position and velocity)
                 [0, 1]]) 

    kf.H = np.array([[1, 0]])  # Measurement function (we observe only count)

    kf.R = np.array([[1000]])  # Measurement noise covariance
    kf.P *= 10000  # Initial state covariance
    kf.Q = np.array([[0.1, 0],  # Process noise covariance
                    [0, 0.1]])  
    kf.x = np.array([[reading[0]], [0]])  # Initial state (count, velocity)

    # Apply Kalman filter to the count data
    filtered_counts = []

    for count in reading:
        kf.predict()
        kf.update([count])
        filtered_counts.append(kf.x[0, 0])  # Store filtered count estimate

    return filtered_counts
```

```python
# New Plant Functions

plt.ylim(0, 75) # Limit range to the dependent range

T60 = pd.read_csv('throughput60IC17.csv') 
T70 = pd.read_csv('throughput70IC25.5.csv')
T80 = pd.read_csv('throughput80IC34.csv')

T60Avg = T60['Filtered'][T60['Filtered']>1].mean()
T70Avg = T70['Filtered'][T70['Filtered']>1].mean()
T80Avg = T80['Filtered'][T80['Filtered']>1].mean()

plt.scatter(T60['Time'], T60['Measured'], color = 'red')
plt.plot(T60['Time'], T60['Filtered'], color = 'red')
plt.plot(T60['Time'], T60Avg * (1-math.e ** (6 * (- T60['Time'] + 6))), color = 'black')
plt.scatter(T70['Time'], T70['Measured'], color = 'green')
plt.plot(T70['Time'], T70['Filtered'], color = 'green')
plt.plot(T70['Time'], T70Avg * (1-math.e ** (6 * (- T70['Time'] + 8.25))), color = 'black')
plt.scatter(T80['Time'], T80['Measured'], color = 'blue')
plt.plot(T80['Time'], T80['Filtered'], color = 'blue')
plt.plot(T80['Time'], T80Avg * (1-math.e ** (6 * (- T80['Time'] + 10.75))), color = 'black')
plt.show()
```
